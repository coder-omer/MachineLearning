```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams["figure.figsize"] = (10,6)
pd.set_option('display.max_columns', 100)
```

```python
df = pd.read_csv("Advertising.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
X = df.drop("sales", axis =1)
y = df["sales"]
```

```python
X.head()
```

### Polynomial Features

```python
from sklearn.preprocessing import PolynomialFeatures
```

We create an artificial overfiting situation by taking poly feature from 5 degrees

```python
polynomial_converter = PolynomialFeatures(degree=5, include_bias=False)

poly_features = polynomial_converter.fit_transform(X)
```

```python
poly_features.shape
```

### Train | Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size=0.3, random_state=101)
```

## Scaling the Data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(X_train)  # Apply just for X_tarin not for X_test (Kalıp çıkarma)
```

```python
X_train_scaled = scaler.transform(X_train) # Apply transform according to fit
X_test_scaled = scaler.transform(X_test)
```

## <span style='color:red'> Elastic net</span> 

```python
from sklearn.linear_model import ElasticNet, ElasticNetCV
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

def train_val(y_train, y_train_pred, y_test, y_pred, name):
    
    scores = {name+"_train": {"R2" : r2_score(y_train, y_train_pred),
    "mae" : mean_absolute_error(y_train, y_train_pred),
    "mse" : mean_squared_error(y_train, y_train_pred),                          
    "rmse" : np.sqrt(mean_squared_error(y_train, y_train_pred))},
    
    name+"_test": {"R2" : r2_score(y_test, y_pred),
    "mae" : mean_absolute_error(y_test, y_pred),
    "mse" : mean_squared_error(y_test, y_pred),
    "rmse" : np.sqrt(mean_squared_error(y_test, y_pred))}}
    
    return pd.DataFrame(scores)
```

```python
elastic_model = ElasticNet(alpha=1, l1_ratio=0.5, random_state=42) # l1_ratio: 1: Lasso or 0:Ridge
elastic_model.fit(X_train_scaled, y_train)

# l1_ratio: Teorideki "fi" diye bahsedilen kısım. 0.5 ile 2 modelide kullanacak eşit oranda
```

```python
y_pred = elastic_model.predict(X_test_scaled)         # Test için predict
y_train_pred = elastic_model.predict(X_train_scaled)  # Train için predict
```

```python
es = train_val(y_train, y_train_pred, y_test, y_pred, "elastic")
es
# Sonuçlar güzel değil önceden 0.98 leri görmüştük
# Default değerleri kullandığımız için sonuçlarımız iyi gelmedi(alpha=1, l1_ratio=0.5)
# Bu sonuçlar doğru sonuçlar mı diye test etmek için CV yapalım altta
```

## <span style='color:red'> Grid Search</span> 

```python
# Gridsearchcv yi tüm ML yöntemleri için uygulayabiliriz 
# Önceki yöntemler o yöntemlere özel "grid search" lerdi diyebiliriz
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
elastic_model = ElasticNet(max_iter=10000, random_state=42)
```

```python
param_grid = {"alpha":[0.01, 0.012, 0.2, 0.5, 0.6, 0.7, 1],
            "l1_ratio":[0.1, 0.5, 0.7, 0.9, 0.95, 0.99, 1]}
# 2 hyperparametrenin alacağı değerleri belirliyoruz önce sonra bunu GridSearchCV nin içine veriyoruz
```

```python
grid_model = GridSearchCV(estimator = elastic_model, param_grid = param_grid, scoring = 'neg_root_mean_squared_error',
                         cv =5, verbose =2)

# estimator  : Kullanılacak model(Hangi model için hyperparametrelere bakacağız)
# verbose =2 : Her 2 iterasyonda 1 bana sonuçları göster
```

```python
grid_model.fit(X_train_scaled, y_train) 
# 0.01-0.1 i 5 kere , 0.01-0.5 i 5 kere, .... 1-0.99 u 5 kere , 1-1 i 5 kere yapıyor
# Yani bütün kombinasyonları deniyor ve en iyi parametreleri buluyor
```

```python
grid_model.best_params_  # En iyi parametreleri grid_search ile bulduk bu sefer(Yukarda da bulmuştuk aynı sonucu)
```

```python
pd.DataFrame(grid_model.cv_results_)
# RMSE değerine göre her bir iterasyondaki en iyi parametrelere karşılık gelen sonuçlar
```

```python
grid_model.best_index_
```

```python
grid_model.best_score_
```

## Using Best Hyperparameters from GridSearch

```python
# En iyi parametreleri kullandığımızda sonuçlarımıza bakalım
```

```python
y_pred = grid_model.predict(X_test_scaled)
y_train_pred = grid_model.predict(X_train_scaled)
```

```python
train_val(y_train, y_train_pred, y_test, y_pred, "GridSearch")
```
