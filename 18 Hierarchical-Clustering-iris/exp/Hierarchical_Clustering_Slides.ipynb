```python
# Hierarchical Clustering
# Unsupervised bir yöntem
# Soldaki şekil gibi bi dendegram çiziyoruz ve kaç sınıfa ayrıldığını göreceğiz
# Burada mavilere baktığımızda 3 cluster a ayrıldığını görüyoruz
# K-means den farkı "k" yı belirlemiyoruz
# Genelde küçük datalarda hierarchical clustering kullanılır. Büyük datalarda yavaş çalışır
```

```python
# Kümelemeyi yaparken smilarity ye(distance a) bakıyor. Yani data noktaları arasındaki uzaklık
# Müşteri 3 e değil de 4 e böl diyebilir. Bu dendogram fikir vermiş oluyor
# Data npktaları birleşerek sınıflar oluşuyor sağ alttaki hocanın çizdiği gibi. En son tek bir grup oluşuyor
```

```python
# Bu algoritma 2 şekilde çalışıyor : 
    # Agglomerative(Kümelemeyi aşağıdan yukarı yapıyor/Tümevarım)
    # Divisive : Böldüğü datayı 2 yer bölüyor sonra tekrar 2 ye bölüyor vs vs en son data noktaları oluşturuyor
        # .. divisive kullanırken içinde K-means algoritması kullanılır
    # NOT: Daha çok agglomerative kullanılır
```

```python
# Clustering deyince
# Minimal intra cluster distance: Clusterlar içindekş mesafenin minimum olmasını istiyoruz
# Maximal inter cluste distance : Clusterlar arasındaki mesafenin maximum olmasını istiyoruz
```

```python
# Feature a göre model nasıl çalışıyor bakalım.
# Model ilk başta her bir datanın birbirine olan uzaklıklarına bakıyor(Sol en üstteki şekil)
# 1: Sonra en yakın 2 datayı belirliyor
# 2: 5-6 birleşti cluster oldu. Sonra bu clusterın diğer noktalara olan uzaklıklarına bakıyor. Bu cluster a en yakın noktayı buluyor
# 3: Sonra o nokta ile o cluster ı birleştiriyor. Sonra yine diğer noktalara olan uzaklıklara bakıyor. Bu cluster a en yakın noktayı buluyor
# 4. Aynı adımlar
# 5. Aynı adımlar
# 6. Aynı adımlar
# 7. En son bu şekil oluşmuş oluyor
# Sonuçta her bir datanın mesafe matrisini hesaplıyor ve buna göre sonuç üretiyor
```
